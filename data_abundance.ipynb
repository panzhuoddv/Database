```python
import pandas as pd
```

```python
data_clean=pd.read_csv('all_superfamily_with_all_genome.csv')#or family
df2=data_clean['number'].groupby([data_clean['genome_name'],data_clean['superfamily']]).sum()
df3=df2.unstack()
df3.fillna(0,inplace=True)
df3=df3.set_index('genome_name')
```

```python
origin_data=pd.read_csv(r'16S.csv')
origin_data=origin_data[origin_data['16S']>=98.6]#16S==98.6 or 97 or 100
```

```python
or_data=origin_data
or_data2=or_data['Cell'].groupby(or_data['Sample']).sum()
data=pd.merge(origin_data,or_data2,left_on='Sample',right_index=True)
data.rename(columns={'Cell_x':'cell','Cell_y':'total_cell_per_sample'},inplace=True)
data2=data.drop_duplicates(['Sample','EMPO3','total_cell_per_sample'])
superfamily_list=list(df3.columns.values)
series_list=[]
for i, superfamily in enumerate(superfamily_list):
    print(i)
    or_data=pd.merge(data,df3[superfamily],how='outer',left_on='Genome',right_index=True)
    or_data[superfamily]=(or_data['cell']*or_data[superfamily])/or_data['total_cell_per_sample']
    N=or_data[superfamily].groupby(or_data['Sample']).sum()
#     or_data[superfamily[0]]=(or_data['cell']*or_data[superfamily[0]])/or_data['total_cell_per_sample']
    series_list.append(N)
    
```

```python
df4=pd.concat(series_list,axis=1)
df_all=pd.merge(data2,df4,how='outer',left_on='Sample',right_index=True)
df_all=df_all.drop(['Seq'],axis=1)
df_all.to_csv('1W_superfamily_16s=98.6.csv')#or 97,100
```
